# KRX Corporate List Fetch
This notebook downloads and reads the KRX corporate list using pandas, adjusting SSL context.

```python
import pandas as pd
import ssl
# configure SSL to bypass verification
ssl._create_default_https_context = ssl._create_unverified_context
```

## Download Corporate List HTML
Define the URL for the KRX corporate list and prepare to fetch it.

```python
url = 'https://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13'
```

## Read HTML into DataFrame
Use pandas.read_html to read the downloaded HTML into a DataFrame.

```python
df = pd.read_html(url)[0]
```

## Display Head and Columns
Print the first few rows of the DataFrame and list its column names.

```python
print(df.head())
print(df.columns)
```
